```python
df_scientist = pd.read_csv(filepath_datascientist)
df_scientist.head(5)
```

```python
df_scientist = df_scientist.drop_duplicates()
```

```python
df_scientist["Salary Estimate"] = df_scientist["Salary Estimate"].str.replace(r'\(Glassdoor est.\)', '', regex=True)
```

```python
df_scientist.head()
```

```python
# 1. Remove '$' and 'K'
df_scientist["Salary Estimate"] = df_scientist["Salary Estimate"].str.replace(r'[\$K]', '', regex=True)
```

```python
df_scientist[['Salary_Min', 'Salary_Max']] = df_scientist["Salary Estimate"].str.split('-', expand=True)
```

```python
df_scientist = df_scientist.drop(columns = "Salary Estimate")
df_scientist.head()
```
