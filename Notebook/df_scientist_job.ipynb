```python
import pandas as pd
import numpy as np
```

```python
df_scientist_job = pd.read_csv("df_scientist_job.csv", index_col=0)
df_scientist_job.head(5)
```

```python
df_scientist_job['Company Name'] = df_scientist_job['Company Name'].str.replace(r'\n\d+(\.\d+)?', '', regex=True)
```

```python
df_scientist_job[['Location_City', 'Location_State']] = df_scientist_job['Location'].str.split(', ', n=1, expand=True)
#df_scientist_job[['Location_City', 'Location_State']] = df_scientist_job['Location'].str.split(', ', expand=True)
```

```python
df_scientist_job = df_scientist_job.drop(columns = "Location")
df_scientist_job.head(5)
```

```python
df_scientist_job[['Headquarters_City', 'Headquarters_State']] = df_scientist_job['Headquarters'].str.split(', ', expand=True)
```

```python
df_scientist_job = df_scientist_job.drop(columns = "Headquarters")
df_scientist_job.head(5)
```

```python
df_scientist_job.columns
```

```python
df_scientist_job = df_scientist_job.drop(columns = "Hourly")
```

```python
df_scientist_job = df_scientist_job.drop(columns = "Employer provided")
```
